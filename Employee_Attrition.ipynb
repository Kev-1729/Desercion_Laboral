```python
def filtrar_outliers(data, columna):
    # Calcular los cuartiles
    Q1 = data[columna].quantile(0.25)
    Q3 = data[columna].quantile(0.75)
    IQR = Q3 - Q1

    # Calcular los límites inferior y superior para los outliers
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    # Filtrar los outliers
    data_filtrada = data[(data[columna] < limite_superior) & (data[columna] >= limite_inferior)]

    return data_filtrada
```

```python
def identificar_outliers(data, columna):
    # Calcular los cuartiles
    Q1 = data[columna].quantile(0.25)
    Q3 = data[columna].quantile(0.75)
    IQR = Q3 - Q1
    # Calcular los límites inferior y superior para los outliers
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    return limite_inferior, limite_superior
```

```python
def fill_missing_values(df, columns):
    # Itera sobre cada columna de la lista
    for column_name in columns:
        # Interpola valores intermedios usando un promedio lineal
        df[column_name] = df[column_name].interpolate(method='linear')

        # Usa forward fill y backward fill para cubrir NaN restantes
        df[column_name].fillna(method='ffill', inplace=True)
        df[column_name].fillna(method='bfill', inplace=True)

        # Redondea a enteros
        df[column_name] = df[column_name].astype(int)

    return df.round(0)
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Crear una instancia de SMOTE
smote = SMOTE(sampling_strategy=0.5, random_state=101)

# Aplicar SMOTE para realizar el sobremuestreo
X_res, y_res = smote.fit_resample(X, y)

# Mostrar la distribución de clases antes y después del sobremuestreo
print("Distribución de clases original:", Counter(y))
print("Distribución de clases después del sobremuestreo:", Counter(y_res))

# Combinar nuevamente las características y la variable objetivo en un DataFrame
df_generado = pd.concat([X_res, y_res], axis=1)
```

```python
from sklearn.feature_selection import VarianceThreshold, f_classif, SelectKBest, chi2

# Definir X e y
X_numeric = df_scalation[numeric_columns]
X_ordinal = df_scalation[ordinal_columns]
X_bool = df_scalation[bool_columns]
y = df_scalation['Attrition'].astype(int)

# 1. Selección de características para Variables Numéricas usando ANOVA F-test
selector_num = SelectKBest(f_classif, k='all')
X_numeric_selected = selector_num.fit_transform(X_numeric, y)
numeric_scores = selector_num.scores_
numeric_pvalues = selector_num.pvalues_

# 2. Selección de características para Variables Ordinales usando ANOVA F-test
selector_ord = SelectKBest(f_classif, k='all')
X_ordinal_selected = selector_ord.fit_transform(X_ordinal, y)
ordinal_scores = selector_ord.scores_
ordinal_pvalues = selector_ord.pvalues_

# 3. Selección de características para Variables Booleanas usando Chi2 y Mutual Information
X_bool = X_bool.astype(int)  # Convertir booleanas a enteros
selector_bool_chi2 = SelectKBest(chi2, k='all')
X_bool_chi2_selected = selector_bool_chi2.fit_transform(X_bool, y)
bool_chi2_scores = selector_bool_chi2.scores_
bool_chi2_pvalues = selector_bool_chi2.pvalues_
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer


# Seleccionar columnas numéricas (int64 y float64)
numeric_columns = X.select_dtypes(include=['float'])

# Rellenar valores faltantes con la media de cada columna
for column in numeric_columns.columns:
    mean_value = numeric_columns[column].mean()
    numeric_columns[column].fillna(value=mean_value, inplace=True)

# Estandarizar las características
scaler = StandardScaler()
df_loan_scaled = scaler.fit_transform(numeric_columns)

# Crear un objeto SimpleImputer para manejar valores faltantes
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mean.fit(df_loan_scaled)

# Aplicar PCA
pca = PCA()
pca.fit(df_loan_scaled)

# Calcular la varianza explicada por cada componente
explained_variance = pca.explained_variance_ratio_

# Calcular la varianza explicada acumulada
cumulative_explained_variance = np.cumsum(explained_variance)
```

```python
# Crear la gráfica
plt.figure(figsize=(10, 6))

# Gráfico de barras para la varianza explicada individual
plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.7, align='center',
        label='Varianza explicada individual', color='skyblue')

# Gráfico de línea para la varianza explicada acumulada
plt.step(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, where='mid',
         label='Varianza explicada acumulada', color='darkred', linewidth=2)

# Configurar etiquetas y título
plt.ylabel('Ratio de varianza explicada')
plt.xlabel('Componentes principales')
plt.legend(loc='best')
plt.title('Varianza Explicada por PCA')

# Añadir cuadrícula
plt.grid(True, linestyle='--', alpha=0.7)

# Mostrar la gráfica
plt.show()
```

```python
# Seleccionar el número de componentes principales
n_components = np.argmax(cumulative_explained_variance >= 0.95)

# Crear un nuevo objeto PCA con el número de componentes seleccionado
pca = PCA(n_components=n_components)

# Aplicar PCA a los datos escalados
df_pca = pca.fit_transform(df_loan_scaled)

# Imprimir el número de componentes seleccionados
print(f"Number of components selected: {n_components}")

# Crear un DataFrame con las componentes principales
df_pca = pd.DataFrame(df_pca, columns=[f'PC{i+1}' for i in range(n_components)])
```
